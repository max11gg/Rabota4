(Обработка датасета:
  - Удаление хотя бы одного признака)
  Функция drop удлаяет признак, указанный в параметрах

```python
df = df.drop(columns=['JP_Sales'])
df
```

(Обработка датасета:
  - Замена значений)
  Функция replace возвращает копию строки, в которой все входные значения заменяются определенными значениями.
  Функция mean вычисляет среднее значение набора значений, переданных в столбцах. А вместе с функцией isnull происходит подсчет в процентах пустых значений.

```python
df = df.replace({'Publisher' :np.nan}, 'MMM')
```

```python
df.isnull().mean()*100
```

(Обработка датасета:
  - Использование или описание простейшего подстановщика)
 Параметр strategy=’mean’ заменяет пропущенные значения, используя среднее значение в каждом столбце.
 Параметр missing_values=np.nan указывает какие занчения необходимо заменять.
 Функция fit устанавливает подстановщик под определенный столбец/признак.
 Функция transform применяет подстановку значений в столбце/признаке.
 

```python
from sklearn.impute import SimpleImputer

simple = SimpleImputer(strategy='mean', missing_values=np.nan)
simple = simple.fit(df[['Year']])
df['Year'] = simple.transform(df[['Year']])
```

(- Выбрать один признак в качестве целевой переменной и обосновать данный выбор)
Функция sort_values применяется для сортировки датафрейма с разлиными параметрами сортировки.
В обосновании можно сказать, что дата выхода игры прямопрапорционально зависит от продаж, чем новее игра, тем больше вероятность ее приобретения.
Поэтому решил отсортировать именно поэтому целевому признаку, чтобы нагляно увидеть самые старые и новейшие игры.

```python
df.sort_values(by='Year')
```

```python
df.sort_values(by='Year',ascending=False)
```

Функция value_counts подсчитывает количестов элементов для каждого значения талицы.
Функция unique находит количестов не повторяющихся нигде параметров.

```python

print("Кол-во представителей компаний", df.Publisher.value_counts())
```

```python
print("Кол-во представителей платформ", df.Platform.value_counts())
print("Кол-во уникальных компаний", len(df.Publisher.unique()))
```
